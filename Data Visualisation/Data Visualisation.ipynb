## Bar-graphs

```python
plt.style.use("seaborn-white")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platinum","Silver"]) # Current Year
plt.bar(x_coordinates+0.25,[20,10,12],width=0.5,label="Next Year") # Next Year
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-2,5)
plt.legend()
plt.show()
```

### Histogram

```python
Xsn = np.random.randn(100)
sigma = 8
u = 70
X = np.round(Xsn*sigma + u)
X2 = np.round(Xsn*7 + 40)

print(X,X2)
```

```python
plt.style.use("seaborn")

plt.hist(X2,alpha=0.8,label="Physics")
plt.hist(X,alpha=0.9,label="Maths")
plt.ylabel("Prob/FreqCount of Students")
plt.xlabel("Marks Range")
plt.title("Histogram")
plt.show()
```

## Data Visualisation Challenge

Given a Movies Dataset, your taskis is to make a visualisation that -
- Plots the Length of Movie Title Names on X-Axis
- Frequency Counts on Y-axis ie Number of Movies having 'x' characters in their title


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read a CSV File
df = pd.read_csv("../2. Working with Libraries/movie_metadata.csv")
```

```python
df.head(n=10)
```

```python
df.columns
```

```python
titles = list(df.get('movie_title'))
```

```python
print(type(titles))
```

```python
print(titles[:5])
```

```python
freq_titles = {}

for title in titles:
    length = len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
#print(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of Movie Title")
plt.ylabel("Freq Count")
plt.title("Movie Data Visualiation Problem")
plt.show()
```
