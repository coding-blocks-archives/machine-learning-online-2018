```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
dfx = pd.read_csv('../Datasets/linearX.csv')
dfy = pd.read_csv('../Datasets/linearY.csv')

x = dfx.values
y = dfy.values

#Normalise 
x2 = (x-x.mean())/x.std()

plt.scatter(x,y,label="Raw Data")
plt.scatter(x2,y,color='orange',label="Normalised Data")
plt.legend()
plt.show()

```
