```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfx = pd.read_csv('Datasets/weightedX.csv')
dfy = pd.read_csv('Datasets/weightedY.csv')

x = dfx.values
y = dfy.values

x = x.reshape((-1,))
y = y.reshape((-1,))

X = (x-x.mean())/x.std()
Y = y
```

```python
x1 = X 
x2 = X**2

X = np.stack((x1,x2),axis=1)
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,0],Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression()
```

## Training

```python
model.fit(X,Y)
```

# Predictions

```python
output = model.predict(X)
```

```python
# Parameters Learnedm
```

```python
bias = model.intercept_
coeff = model.coef_

print(bias)
print(coeff)
```

## Score 

```python
model.score(X,Y)
```

```python
## Visualise

plt.scatter(X[:,0],Y,label='data')
plt.scatter(X[:,0],output,color='orange',label='prediction')
plt.legend()
plt.show()
```
